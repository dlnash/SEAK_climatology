### Import WRF data

```python
# get list of filenames that contain data from that year from current year folder
filenames = []
for name in glob.glob(path_to_work + 'downloads/wrf-AK/WRFDS_2018-01*.nc'):
    filenames.append(name)
# sort filenames so they are in chronological order
filenames = sorted(filenames)

# read files and extract accumulated precipitation
wrf = preprocess_PCPT(filenames)
wrf
```

```python
diff_lat = wrf.lat.values[2] - wrf.lat.values[1]
diff_lon = wrf.lon.values[2] - wrf.lon.values[1]
### pull precip time series from data
# select the grid cell closest to each of the communities
df_lst2 = []
row_lbl2 = []
for i, (slon, slat) in enumerate(zip(xs, ys)):
    # # select nearest grid cell to station
    # ds = wrf.sel(lat=slat, lon=slon, method="nearest")
    # df = ds.prec.to_dataframe()
    # df_lst2.append(df)
    
    ## TODO: select 9 nearest grid cells
    ds = wrf.sel(lat=slice(slat-diff_lat*1.5, slat+diff_lat*1.5), lon=slice(slon-diff_lon*1.5, slon+diff_lon*1.5))
    ds = ds.mean(("lon", "lat"))
    df = ds.prec.to_dataframe()
    df_lst2.append(df)
    
    # make nice labels for plot
    lbl = u"{:.2f}\N{DEGREE SIGN}N, {:.2f}\N{DEGREE SIGN}W".format(slat, slon*-1)
    row_lbl2.append(lbl)

# df_lst2[0]
```

### Read ERA5 IVT data

```python
datelst = ['0111', '0112', '0113', '0114', '0115', '0116']
filenames = []
for i, monthday in enumerate(datelst):
    filenames.append('/data/downloaded/Reanalysis/ERA5/IVT/2018/ERA5_IVT_2018{0}.nc'.format(monthday))
    
def preprocess(ds):
    '''keep only selected lats and lons'''
    return ds.sel(lat=slice(ext1[2], ext1[3]), lon=slice(ext1[0], ext1[1]))

era = xr.open_mfdataset(filenames, preprocess=preprocess, combine='by_coords')

# select the grid cell closest to each of the communities
df_lst = []
row_lbl = []
rank = []
for i, (slon, slat) in enumerate(zip(xs, ys)):
    ds = era.sel(lat=slat, lon=slon, method="nearest")
    df = ds.IVT.to_dataframe()
    df_lst.append(df)
    
    # calculate AR rank
    rank.append(AR_rank(df))
    
    # make nice labels for plot
    lbl = u"{:.2f}\N{DEGREE SIGN}N, {:.2f}\N{DEGREE SIGN}W".format(slat, slon*-1)
    row_lbl.append(lbl)

df_lst[0]
```

### Read and merge ASOS data

```python
## location and name of txt file
filename = path_to_data + 'downloads/asos_jan2018.csv'
df = pd.read_csv(filename, header=0, sep=',', engine='python',
                      parse_dates=['valid'])

# Converting the index as date (need to do this to utilize functionality of resample)
df.index = pd.to_datetime(df.valid)
stn_lst = df.station.unique()
# # list column names
# for col in df.columns:
#     print(col)

# p01i:
# one hour precip in inches
# for the period from the observation time to the time of the previous hourly precipitation reset

merge_df_lst = []
for i, stn in enumerate(stn_lst):    
    ## select single station
    idx = df['station'] == stn
    tmp = df[idx]

    ## convert from inches/hour to mm/hour
    tmp.loc[:,'p01i'] = tmp['p01i']*25.4

    ## resample to hourly precipitation
    precip_1hour = tmp['p01i'].resample('1H').max()

    tmp2 = precip_1hour.to_frame()
    tmp2 = tmp2.reset_index() # reset index to merge with IVT
    tmp2 = tmp2.rename(columns={"valid": "time"})
    
    # IVT data
    tmp3 = df_lst[i].reset_index()
    
    merged_data = pd.merge(tmp3, tmp2, how='outer', on='time')
    merge_df_lst.append(merged_data)
    
merge_df_lst[1]
```

### Merge dfs to single df

```python
merge_df_final = []
for i, (df1, df2) in enumerate(zip(merge_df_lst, df_lst2)):    
    
    tmp2 = df2.reset_index()
    tmp2 = tmp2.rename(columns={"lat": "wrflat", "lon": "wrflon"})
    
    
    merged_data = pd.merge(df1, tmp2, how='outer', on='time')
    merge_df_final.append(merged_data)
    
merge_df_final[0]
```

## Plot time series

```python
# Set seaborn plot style
sns.set()
sns.set_style("ticks", {'patch.force_edgecolor':False})

### Get x-tick and x-labels
df = merge_df_lst[1]
x = df.time.values
x_ticks = x

x_labels = []
for j in range(len(x)):
    t = pd.to_datetime(str(x[j]))
    x_labels.append(t.strftime('%m-%d %H'))
```

```python
fname = path_to_figs + 'wrfprecip-ERA5IVT_time_series_jan2018_nearneighbor'
fmt = 'png'
nrows = 3
ncols = 2
nplots=nrows*ncols

xaxis_lbl = ['']*4 + ['Date (month-day hour)']*2
ylbl = ['IVT (kg m$^{-1}$ s$^{-1}$)', '']*3
ylbl2 = ['', 'Precipitation (mm hr$^{-1}$)']*3

fig = plt.figure(figsize=(15,15))
sns.set_style("whitegrid")

for i, (df, lbl) in enumerate(zip(merge_df_final, lbl1)):
    
    ax = fig.add_subplot(nrows, ncols, i+1)
    
    # plot IVT
    ivt = sns.lineplot(x=df.index, y="IVT", data=df, color='tab:blue', ax=ax, label='IVT', legend=False)
    
    # add 250 kg m-1 s-1 threshold
    ax.axhline(y=250.0, color='tab:red', linestyle='--')

    
    ax.set_ylabel(ylbl[i])
    ax.set_ylim(0, 1000)

    ax.set_title(lbl+' ({0})'.format(row_lbl[i]), loc='left')
    ax.set_title('AR Scale {0}'.format(rank[i]), loc='right')
    
    ## add twin axis for precip info
    ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis
    prec = sns.barplot(x=df.index, y="prec", data=df, color='k', alpha=0.5, ax=ax2) # wrf-precip
    prec2 = sns.barplot(x=df.index, y="p01i", data=df, color='tab:blue', alpha=0.7, ax=ax2) # asos-precip
    ax2.set_ylabel(ylbl2[i])
    ax2.set_ylim(0, 15)
    
    ## x-axis format dates/ticks
    ax2.set_xticks(x_ticks[::24]) # set the labels every 24 hours
    ax2.set_xticklabels(x_labels[::24], fontsize=10)
    ax2.set_xlim(df.index.min(), df.index.max())
    ax2.set_xlabel(xaxis_lbl[i])
    

# Save the figure
fig.savefig('%s.%s' % (fname, fmt), bbox_inches='tight', dpi=300)

# Show
plt.show()


```

### Plot Storm Total Precipitation

```python
# Set up projection
datacrs = ccrs.PlateCarree()  ## the projection the data is in
mapcrs = ccrs.Mercator() ## the projection you want your map displayed in

# Set tick/grid locations
lats = wrf.lat.values
lons = wrf.lon.values
dx = np.arange(lons.min().round(),lons.max().round()+1,1)
dy = np.arange(lats.min().round(),lats.max().round()+1,1)
```

```python
storm_total_prec = wrf.sum('time')
print('Max storm total:', storm_total_prec.max().values)

# Create figure
fig = plt.figure(figsize=(8, 12))
fig.dpi = 300
fname = path_to_figs + 'jan2018_prec_total'
fmt = 'png'

nrows = 1
ncols = 1

# Set up Axes Grid
axes_class = (GeoAxes,dict(projection=mapcrs))
axgr = AxesGrid(fig, 111, axes_class=axes_class,
                nrows_ncols=(nrows, ncols), axes_pad = 0.45,
                cbar_location='right', cbar_mode='single',
                cbar_pad=0.05, cbar_size='3%',label_mode='')


for k, ax in enumerate(axgr):
    ax = draw_basemap(ax, extent=ext1, xticks=dx, yticks=dy,left_lats=True, right_lats=False)

    # Contour Filled
    prec = storm_total_prec.prec.values
    cf = ax.contourf(lons, lats, prec, transform=datacrs,
                     levels=np.arange(2, 375, 25), cmap=nclc.cmap('WhiteBlueGreenYellowRed'), alpha=0.9, extend='max')


# Colorbar (single)
cb = fig.colorbar(cf, axgr.cbar_axes[0], orientation='vertical', drawedges=False)
cb.set_label('precipitation (mm)', fontsize=11)
cb.ax.tick_params(labelsize=10)

fig.savefig('%s.%s' %(fname, fmt), bbox_inches='tight', dpi=fig.dpi)

# Show
plt.show()
```
