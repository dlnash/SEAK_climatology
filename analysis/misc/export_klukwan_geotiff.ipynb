```python
# Standard Python modules
import os, sys
import xarray as xr
import rioxarray
```

```python
path_to_data = '/work/dnash/data/'
path_to_out = '/data/projects/Comet/cwp140/preprocessed/'

fname = path_to_data + 'preprocessed/SEAK-WRF_PCPT_daily_composite.nc'
prec_mean = xr.open_dataset(fname)

prec_mean = prec_mean.sel(community='Klukwan')
prec_mean.rio.to_raster(path_to_out+"SEAK-WRF_PCPT_daily_composite_mean_klukwan.tif")
```

```python
fname = path_to_out+'SEAK-WRF_PCPT_daily_composite_max.nc'
prec_max = xr.open_dataset(fname)
prec_max = prec_max.sel(community='Klukwan')

prec_max.rio.to_raster(path_to_out+"SEAK-WRF_PCPT_daily_composite_max_klukwan.tif")
```
