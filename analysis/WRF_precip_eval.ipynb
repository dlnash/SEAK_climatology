```python
import sys
import boto3
import botocore
import requests
import netCDF4 as nc
import xarray as xr
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

# import personal modules
# Path to modules
sys.path.append('../modules')
# Import my modules
from AWS_funcs import get_s3_keys
```

```python
bucket_name = 'wrf-se-ak-ar5'
product_name = 'cfsr'
spatial_res = '4km'
temporal_res = 'hourly'
year = 2018
month = 1
start_day = 10
end_day = 16

prefix = product_name + '/' + spatial_res + '/' + temporal_res + '/' + str(year)  \
+ '/' + 'WRFDS_' + str(year) + '-' + str(month).zfill(2) 

```

```python
# No AWS keys required
client = boto3.client('s3', config=botocore.client.Config(signature_version=botocore.UNSIGNED))

paginator = client.get_paginator('list_objects')
result = paginator.paginate(Bucket=bucket_name, Delimiter='/')
for prefix in result.search('CommonPrefixes'):
    print(prefix.get('Prefix'))
```

```python
keys = get_s3_keys(bucket_name, prefix = prefix)
# selecting the day of the month to download -1 to account for zero indexing
s3_data_key = [key for key in keys][start_day-1] 

## download netcdf for specified key
resp = requests.get('https://' + bucket_name + '.s3.amazonaws.com/' + s3_data_key)
```

```python
file_name = s3_data_key.split('/')[-1].split('.')[0]
nc4_ds = nc.Dataset(file_name, memory = resp.content)
store = xr.backends.NetCDF4DataStore(nc4_ds)
```

```python
# open recently downloaded DS
DS = xr.open_dataset(store)
DS
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

DS.PCPT.isel(Time=0).plot.contourf(ax=ax, transform=ccrs.PlateCarree())
ax.coastlines()
```
